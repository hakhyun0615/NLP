# 12-01 케라스를 이용한 태깅 작업 개요(Tagging Task using Keras)

태깅 작업(Tagging Task): 지도 학습(Supervised Learning)  
개체명 인식(Named Entity Recognition): 각 단어의 유형이 사람, 장소, 단체 등 어떤 유형인지 파악  
품사 태깅(Part-of-Speech Tagging): 각 단어의 품사가 명사, 동사, 형용사 인지를 파악

## 훈련 데이터에 대한 이해

X: 태깅을 해야하는 단어 데이터  
y: 레이블에 해당되는 태깅 정보 데이터
