# 11-07 문자 임베딩(Character Embedding)

워드 임베딩과는 다른 방법으로 단어의 벡터 표현 방법을 얻는 문자 임베딩(Character Embedding)에 대해서 알아보겠습니다. 문자 임베딩은 워드 임베딩의 대체재로서 쓰거나, 문자 임베딩을 워드 임베딩과 연결(concatenate)하여 신경망의 입력으로 사용하기도 합니다.

## 1D CNN을 이용한 문자 임베딩

1D CNN: 전체 시퀀스 입력 안의 더 작은 시퀀스에 집중하여 정보를 얻어내는 동작을 하는 알고리즘  
문자 임베딩에 1D CNN 사용할 경우: 문자의 N-gram으로부터 정보를 얻음
  
이렇게 단어 벡터를 얻을 경우, 어떤 단어이든 기본적으로 문자 레벨로 쪼개므로 기존 워드 임베딩의 접근에서 OOV라고 하더라도 벡터를 얻을 수 있습니다. 가령, 'docker'라는 영단어가 훈련 데이터에 없었으나 테스트 데이터에 존재하는 단어였다고 해봅시다. Word2Vec이나 GloVe의 경우에는 OOV 문제가 발생하게 되겠지만, 1D CNN을 이용하는 경우에는 'd', 'o', 'c', 'k', 'e', 'r'로 전부 분리되어 각 문자로 임베딩이 되고나서 1D CNN을 거친 후에 'docker'의 벡터를 얻게됩니다.
